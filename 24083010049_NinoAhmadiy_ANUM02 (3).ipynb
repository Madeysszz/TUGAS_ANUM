```python
#interval
x = np.linspace(0, 10, 500)
```

```python
# deret fungsi sinusoidal
f = np.sin(x) + 0.5*np.sin(2*x) +  (1/3)*np.sin(3*x) + (1/4)*np.sin(4*x) + (1/5)*np.sin(5*x)
```

*NOMOR 1*

```python
#hitung integral dan diferensial dari fungsi 'f' untuk interval x

# H (membagi interval 0 sampai 10 menjadi 500 titik yang jaraknya sama besar)
h = x[1] - x[0]

# forward difference
f_forward = (f[1:] - f[:-1]) / h

#backward difference
f_backward = (f[1:] - f[:-1]) / h

#central difference
f_central = (f[2:] - f[:-2]) / (2*h)

x_forward = x[:-1]
x_backward = x[1:]
x_central = x[1:-1]

print("Panjang Forward Difference :", len(f_forward))
print("Panjang Backward Difference:", len(f_backward))
print("Panjang Central Difference :", len(f_central))
```

Code ini melakukan perhitungan turunan numerik menggunakan metode forward difference, backward difference, dan central difference pada deret fungsi sinusoidal yang didefinisikan dalam interval dari 0 sampai 10 dengan 500 titik, kemudian menghitung panjang array hasil dari setiap metode diferensiasi tersebut untuk memastikan konsistensi ukuran data pada proses diferensiasi numerik.

```python
plt.figure(figsize=(9,5))
plt.plot(x, f, 'k-', label="f(x)", linewidth=1.2, marker='o', markersize=2.5)
plt.plot(x_forward, f_forward, color="orange", linestyle="--", label="Forward difference", linewidth=1, marker='s', markersize=2)
plt.plot(x_backward, f_backward, color="teal", linestyle=":", label="Backward difference", linewidth=1, marker='^', markersize=2.5)
plt.plot(x_central, f_central, color="purple", linestyle="-.", label="Central difference", linewidth=1, marker='d', markersize=2.5)

plt.title("Pendekatan Turunan Fungsi f(x) dengan Metode Numerik", fontsize=12)
plt.xlabel("x", fontsize=10)
plt.ylabel("f(x)", fontsize=10)
plt.legend(loc="best", fontsize=8)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

Grafik ini menunjukkan pendekatan turunan fungsi $ f(x) $ menggunakan metode numerik berupa forward difference, backward difference, dan central difference, yang dibandingkan dengan grafik fungsi aslinya. Terlihat bahwa metode central difference (ungu), forward difference (oranye), dan backward difference (hijau kebiruan) memberikan hasil turunan numerik yang sangat mirip dan mengikuti pola perubahan fungsi, sedangkan fungsi aslinya (hitam) berbeda karena bukan turunan. Grafik ini mengilustrasikan bagaimana metode numerik tersebut dapat digunakan untuk mendekati turunan fungsi secara efektif pada berbagai titik $x$.

*NOMOR 2*

```python
#  hitung secara terpisah bagian integral di atas dan di bawah y=0

F_total = np.cumsum((f[:-1] + f[1:]) / 2 * h)   
F_total = np.insert(F_total, 0, 0)

print("Nilai integral total :", F_total[-1])

plt.figure(figsize=(9,5))
plt.plot(x, F_total, color="darkgreen", linestyle="--", linewidth=1.2,
         marker='o', markersize=3, label="Integral f(x)")
plt.fill_between(x, F_total, color="salmon", alpha=0.2) 
plt.title("Hasil Integral Numerik f(x)", fontsize=13, fontweight="bold")
plt.xlabel("x", fontsize=11)
plt.ylabel("∫ f(x) dx", fontsize=11)
plt.legend(fontsize=9, loc="upper left")
plt.grid(True, linestyle=":", alpha=0.6)
plt.tight_layout()
plt.show()
```

Hasilnya adalah visualisasi numerik dari integral fungsi $ f(x) $, dengan nilai total integral sekitar 2.1864 yang ditampilkan di konsol. Grafik ini memudahkan pemahaman tentang bagaimana integral numerik dari data diskret terakumulasi sepanjang interval.

*NOMOR 3*

```python
# Pisahkan area atas & bawah
f_upper = np.clip(f, 0, None)
f_lower = np.clip(f, None, 0)

area_upper = np.trapz(f_upper, x)
area_lower = np.trapz(f_lower, x)

print("Luas integral di atas sumbu x  :", area_upper)
print("Luas integral di bawah sumbu x :", area_lower)

# Plotting 2 subplot
fig, axs = plt.subplots(2, 1, figsize=(9,7))

# --- Subplot 1: f(x) biasa
axs[0].plot(x, f, color="blue", linewidth=1.3, marker="o", markersize=3, label="f(x)")
axs[0].axhline(0, color="black", linewidth=0.8, linestyle="--")
axs[0].set_title("Fungsi f(x)")
axs[0].legend(); axs[0].grid(True, linestyle=":")

# --- Subplot 2: area atas & bawah
axs[1].plot(x, f, color="darkorange", linewidth=1.3, label="f(x)")
axs[1].fill_between(x, f_upper, 0, color="lightgreen", alpha=0.6, label="Area atas (+)")
axs[1].fill_between(x, f_lower, 0, color="salmon", alpha=0.6, label="Area bawah (-)")
axs[1].axhline(0, color="black", linewidth=0.8)
axs[1].set_title("Integral f(x) → area positif & negatif")
axs[1].legend(); axs[1].grid(True, linestyle=":")

plt.tight_layout()
plt.show()
```

Visualisasi ini menampilkan grafik fungsi f(x) yang dibagi menjadi dua bagian utama, yaitu area di atas sumbu x (area positif) dan area di bawah sumbu x (area negatif). Bagian atas grafik, yang diarsir berwarna kuning, merepresentasikan integral positif f(x) sebagai luas area di atas sumbu x, sedangkan bagian bawah yang diarsir warna merah menunjukkan integral negatif sebagai luas area di bawah sumbu x. Dengan menggunakan metode trapezoidal, luas masing-masing area dihitung untuk memberikan pemahaman yang jelas tentang kontribusi nilai fungsi terhadap integral total pada sumbu x. Grafik ini memperlihatkan bagaimana fungsi berubah dari positif ke negatif, serta mempertegas pentingnya memisahkan daerah integral agar interpretasi luas area lebih informatif dan visualisasi integral menjadi lebih intuitif.
