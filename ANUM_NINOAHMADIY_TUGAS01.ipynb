```python
# NAMA : NINO AHMADIY
# NPM : 24083010049
# KELAS : ANALISIS NUMERIK (A)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

x = np.linspace(3, 10, 500)
y = skewnorm.pdf(x, a=6, loc=3.5, scale=1.5) * 10
np.random.seed(49)  # masukkan NRP
y = y + np.random.normal(0, 0.02, size=x.shape)  # add Gaussian
hargaproduksi = 0.1 * x + 3
```

```python
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='navy', linewidth=3)
plt.plot(x, hargaproduksi, color='red', linewidth=2)
plt.title('Penjualan')
plt.xlabel('x (harga, ribuan)')
plt.ylabel('y (jumlah barang laku, ratusan)')
plt.grid(True)
plt.show()
```

Hubungan Penjualan, saat harga (dari kanan ke kiri), awalnya jumlah barang yang laku awalnya meningkat, mencapai puncaknya, lalu menurun dikarenakan harga yang semakin tinggi, lalu garis merah yang ada di visualisasi tersebut mengindikasikan trend harga produksi yang dapat membantu menganalisis marginal cost atau strategi dalam menaruh harga yang baik.

```python
#keuntungan penjualan
keuntungan = (x - hargaproduksi) * y * 100  
```

```python

idx_max = np.argmax(keuntungan)
x_opt = x[idx_max]             # harga optimal
y_opt = y[idx_max] * 100       # jumlah barang laku optimal
max_profit = keuntungan[idx_max]

print(f"Harga optimal       : {x_opt:.2f} ribuan")
print(f"Jumlah barang laku : {y_opt:.0f} unit")
print(f"Keuntungan maksimum: {max_profit:.2f} ribu")
```

Program ini berfungsi untuk menentukan harga jual dan jumlah barang yang terjual untuk memaksimalkan keuntungan dalam konteks penjualan.

```python
step = 50
harga_jual = np.round(x[::step], 2)
harga_produksi = np.round(hargaproduksi[::step], 2)
jumlah_laku = np.round(y[::step] * 100, 0)
profit = np.round(keuntungan[::step], 2)

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, y, color='navy', linewidth=2, label='Jumlah barang laku (ratusan)')
ax.plot(x, hargaproduksi, color='red', linewidth=2, label='Harga produksi (ribuan)')
ax.plot(x, keuntungan/1000, color='green', linewidth=2, label='Keuntungan (ribuan/1000)')
ax.axvline(x_opt, color='black', linestyle='--', label=f'Harga optimal = {x_opt:.2f}')

ax.set_title("Plot Penjualan, Harga Produksi, dan Keuntungan")
ax.set_xlabel("Harga jual (ribuan)")
ax.set_ylabel("Nilai")
ax.grid(True)
ax.legend()


col_labels = ["Harga Jual", "Harga Produksi", "Jumlah Laku", "Keuntungan"]
table_data = list(zip(harga_jual, harga_produksi, jumlah_laku, profit))

table = plt.table(cellText=table_data, colLabels=col_labels,
                  loc='bottom', cellLoc='center', bbox=[0, -0.5, 1, 0.4])

plt.subplots_adjust(left=0.1, bottom=0.35)  # kasih ruang buat tabel
plt.show()
```

dari grafik, terlihat bahwa saat harga mendekati titik optimal, jumlah barang yang terjual mencapai puncaknya, dan keuntungan juga meningkat sebelum akhirnya menurun kembali.Keuntungan paling tinggi tercatat saat harga belum terlalu tinggi, menunjukkan adanya batasan harga di mana permintaan mulai menurun dan mempengaruhi total penjualan.

```python
plt.figure(figsize=(8,5))
plt.plot(x, keuntungan, color='black', linewidth=2)
plt.title('Keuntungan')
plt.xlabel('x (harga, ribuan)')
plt.ylabel('Keuntungan (ribu)')
plt.grid(True)
plt.axvline(x_opt, color='red', linestyle='--', label='Harga optimal')
plt.legend()
plt.show()
```

Garis putus-putus menunjukkan harga optimal yang berada di angka 4.91. Ini adalah harga di mana keuntungan maksimal dapat dicapai, jadi kesimpulan yang dapat kita tarik dalam visualisasi ini adalah memahami titik optimal dalam penetapan harga sangat penting untuk memaksimalkan keuntungan, serta mengatur strategi penjualan agar tetap kompetitif di pasar.
